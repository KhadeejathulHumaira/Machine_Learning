```python
from matplotlib import pyplot as plt
import pandas as pd
df=pd.read_csv("loan.csv")
df.head(15)

```

```python
plt.style.use('seaborn-whitegrid')

fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=df["Income"],
                  y=df["Savings"],
                  c=df["Classification"],
                  cmap="winter")
#Customize the plot

ax.set(xlabel="Income",
      ylabel="Savings");
#Legend
ax.legend(*scatter.legend_elements(),title="Chance");
ax.axhline(y=8000,
          linestyle='--');
ax.axvline(x=10000,
          linestyle='--');
fig.suptitle("Loan Analysis",fontsize=16,fontweight="bold");
```

```python
fig.savefig("loan-analysis-image.png")
```
